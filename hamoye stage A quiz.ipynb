```python
import numpy as np
import pandas as pd
url = 'https://raw.githubusercontent.com/WalePhenomenon/climate_change/master/fuel_ferc1.csv?raw=true'
fuel_Data = pd.read_csv(url, error_bad_lines=False)
fuel_Data.describe(include='all')
```

```python
fuel_Data.groupby('fuel_type_code_pudl')['fuel_type_code_pudl'].count()
```

```python
fuel_Data.isna().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize =(7,4))
plt.xticks(rotation =90)
sns.barplot(data=fuel_Data, x= 'report_year', y= 'fuel_cost_per_unit_delivered')
plt.xlabel('average fuel cost per unit delivered')
plt.show()
```
